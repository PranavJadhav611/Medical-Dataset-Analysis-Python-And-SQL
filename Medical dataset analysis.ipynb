# Task 1: Loading Hospitalization Details

```python
import pandas as pd
```

```python
file_path = "P:/PortFolio Projects/Medical Dataset Analysis Project/hospitalisation_details.csv"
hospital_data = pd.read_csv(file_path)
```

# Task 2: Identifying Null Values in Hospitalization Details

```python
null_values = hospital_data.isnull().sum()
print("\nNull Values:\n", null_values)
```

```python
with_na = hospital_data[hospital_data['mth'].isna()]
with_na
```

```python
hospital_data['mth']= hospital_data['mth'].fillna(value='None')                       # Filled the null values with None
```

```python
hospital_data.head()
```

# Task 3: Identifying Data Types in Hospitalization Details

```python
data_types = hospital_data.dtypes
print("\nData Types:\n", data_types)
```

# Task 4: Identifying Duplicate Data in Hospitalization Details

```python
len(hospital_data.index)                 # Number of rows Before removing duplicates
```

```python
duplicates = hospital_data[hospital_data.duplicated()]
duplicates
```

# Task 5: Data Preprocessing and Cleaning for Hospitalization Detail

```python
hospital_data = hospital_data.drop_duplicates()
```

```python
len(hospital_data.index)           # Number of rows After removing duplicates
```

```python
len(hospital_data.columns)    # Number of columns Before droping columns.
```

```python
hospital_data = hospital_data.drop(['Has_Children', 'Is_Frequent_Treatment'], axis=1)
```

```python
len(hospital_data.columns)    # Number of columns After droping columns.
```

```python
hospital_data.head()
```

```python
hospital_data = hospital_data.rename(columns={'c_id': 'Client_ID', 'yr': 'Year', 'mth':'Month' ,'date?':'Date' , 'children?': 'Number_of_children ' , 'charges?': 'Charges'})
hospital_data.head()
```

```python
hospital_data.to_csv('hospitalisation_details_cleaned.csv', index=False)       # Saved cleaned file
```
